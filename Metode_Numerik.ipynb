```python
from pylab import *
import numpy as np
x = [0., 1., 2., 3., 4., 5., 6., 7., 8., ]
y = [1., 1.3, 2.4, 3.1, 3.9, 4.8, 6.1, 6.7, 8.]

n = len(x)

###
def L(i, t):
    global x
    p = 1
    for j in range (n):
        if (j!=i):
            p = p* (t-x[j])/(x[i] -x[j])
    return p

def f(t):
    global x, y
    s = 0
    for i in range (n):
        s = s + L(i, t)*y[i]

    return s
###

t = linspace (-1, 9, 100)
z = f(t)

grid(True)
ylim (-1, 11)

plot (t, z)

scatter(x, y, s = 100, c = np.random.rand(n), alpha = .5)
show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 20, 40, 60, 80, 100],float)
y = np.array([26.0, 48.6, 61.6, 71.2, 74.8, 75.2],float)

xplot = np.linspace(x[0],x[-1])
yplot = np.array([],float)

for xp in xplot:
    yp = 0

    for xi,yi in zip(x,y):
        yp += yi * np.prod((xp - x[x != xi])/(xi - x[x != xi]))
    yplot = np.append(yplot,yp)

#plot
plt.plot(x, y, 'bo', xplot,yplot, 'r-')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# metode simpson 1/3

#define function to integrate
def f(x):
  return 1/(1 + x)

#implementing simpson's 1/3
def simpson13(x0, xn, n):
  h = (xn - x0) / n
  integration = f(x0) + f(xn)

  for i in range(1,n):
    k = x0 + 1*h

    if i%2 == 0:
      integration = integration + 2 * f(k) #untuk ganjil
    else:
      integration = integration + 4 * f(k) #untuk genap

  #hasil nilai integration
  integration = integration * h/3
  return integration

lower_limit = float(input("Masukan Batas Bawah: "))
upper_limit = float(input("Masukan Batas Atas: "))
sub_interval = int(input("Masukan Banyaknya iterasi: "))

#print hasilnya
result = simpson13(lower_limit, upper_limit, sub_interval)
print("Hasil integrasi simpson's 1/3nya adalah: %0.6f" % (result))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def func(y,v,t):
    return v - 3*y + t

t0 = 0.0
tn = 4.0
y0 = 1.0
v0 = -2.0
ndata = 100
t = np.linspace(t0,tn,ndata)

y = np.zeros(ndata)
y[0] = y0
v = np.zeros(ndata)
v[0] = v0
h = t[2] - t[1]

for i in range(0,ndata-1):
    y[i+1] = y[i] + h*v[i]
    v[i+1] = v[i] + h*func(y[i],v[i],t[i])

plt.figure(1)
plt.title('Y Euler utk ODE-2')
plt.plot(t,y)
plt.legend()

plt.figure(2)
plt.title('V Euler utk ODE-2')
plt.plot(t,v)
plt.show()
```

```python
import math
import matplotlib.pyplot as plt

dy = lambda x,y: 2*y
f = lambda x: math.exp(2*x)

xi = 0; xf = 1; h = 0.50
n = int((xf - xi) / h)
x = 0; y = 1

print('x \t\t y'); print('%f \t %f'% (x,y))
x_plot = []; y_euler = []; y_analytical = []

#Metode Euler
for i in range(1, n+1):
  x_plot.append(x); y_euler.append(y); y_analytical.append(f(x))
  y = y + dy(x, y) * h
  x = x + h
  print('%f \t %f'% (x,y))

x_plot.append(x); y_euler.append(y); y_analytical.append(f(x))
plt.plot(x_plot, y_euler, x_plot, y_analytical)
plt.legend(['Euler', 'Analytical'])
```
